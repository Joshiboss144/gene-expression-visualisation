```python
#Basic libraries we learned in Year in CS
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
os.makedirs("../results", exist_ok = True)
```

```python
# --- Load ---
file_path = "../data/breast_cancer_dataset.csv"
df = pd.read_csv(file_path)  # Load the breast cancer dataset into a pandas DataFrame

# --- Clean ---
# Drop any empty/unnamed columns
for c in list(df.columns):
    if str(c).lower().startswith("unnamed"):
        df = df.drop(columns=[c])  # Remove columns with names starting with "unnamed"

# Map labels to readable text + simple numeric label
if "diagnosis" in df.columns:
    df["diagnosis"] = (
        df["diagnosis"].astype(str).str.strip().str.upper().map({"B": "Benign", "M": "Malignant"})
        # Convert diagnosis codes: B -> Benign, M -> Malignant
        # Steps: convert to string, remove whitespace, uppercase, then map values
    )
else:
    # If no diagnosis columns, create a dummy so code still runs
    df["diagnosis"] = np.nan  # Create placeholder with NaN values if diagnosis column doesn't exist

df["diagnosis_num"] = df["diagnosis"].map({"Benign": 0, "Malignant": 1})  # Create numeric encoding: Benign=0, Malignant=1

# Numeric features only (skip obvious id/label columns)
numeric_cols = [
    c for c in df.columns
    if pd.api.types.is_numeric_dtype(df[c]) and c not in ["id", "diagnosis_num"]
    # Create list of numeric feature columns, excluding ID and diagnosis_num columns
]
```

```python
# Helper: choose a feature if preferred one is missing
def pick_feature(preferred: str, fallback_list):
    # Check if preferred feature exists and is numeric
    if preferred in df.columns and pd.api.types.is_numeric_dtype(df[preferred]):
        return preferred
    # Try each fallback feature in order
    for f in fallback_list:
        if f in df.columns and pd.api.types.is_numeric_dtype(df[f]):
            return f
    # If no preferred or fallback features are available, use the numeric feature with highest variance
    if numeric_cols:
        return df[numeric_cols].var().sort_values(ascending=False).index[0]
    # If no numeric columns at all, raise an error
    raise ValueError("No numeric columns available for plotting.")

# Helper: safe boxplot that won't crash when groups are empty
def safe_boxplot(data, x, y, title, out_path):
    # Remove rows with NaN values in the x or y columns
    sub = data[[x, y]].dropna()
    # Count the number of samples in each group
    if x in sub.columns:
        counts = sub[x].value_counts(dropna=False).to_dict()
    else:
        counts = {}
    # Skip plotting if there's no data after removing NaNs
    if len(sub) == 0 or sub[y].dropna().empty or sub[x].dropna().empty:
        print(f"[SKIP] {title}: nothing to plot (after dropna). Counts: {counts}")
        return
    # Create and save the boxplot
    plt.figure(figsize=(6,5))
    sns.boxplot(data=sub, x=x, y=y)
    plt.title(title)
    plt.tight_layout(); plt.savefig(out_path, dpi=300); plt.show()
```

```python
# Plot 1: Histogram (radius_mean)
feat_hist = pick_feature("radius_mean", ["perimeter_mean", "area_mean", "texture_mean"]) 
plt.figure(figsize=(8,5))
sns.histplot(df[feat_hist].dropna(), bins=30, kde=True)
plt.title(f"Distribution of {feat_hist}")
plt.xlabel(feat_hist)
plt.ylabel("count")
plt.tight_layout(); plt.savefig("../results/plot1_hist_radius_mean.png", dpi=300); plt.show()
```

```python
# Plot 2: Boxplot (feature by diagnosis)
feat_box = pick_feature("area_mean", ["radius_mean", "perimeter_mean", "texture_mean"]) 
# Only attempt if we have a diagnosis column with some labels
if df["diagnosis"].notna().any():
    safe_boxplot(df, x="diagnosis", y=feat_box,
                 title=f"{feat_box} by Diagnosis",
                 out_path="../results/plot2_box_area_mean_diagnosis.png")
else:
    print("[SKIP] Boxplot: 'diagnosis' column is missing or all NaN.")
```

```python
# Plot 3: Top-10 correlations with malignancy
if df["diagnosis_num"].notna().any() and len(numeric_cols) > 0:
    corr_with_y = df[numeric_cols].corrwith(df["diagnosis_num"]).dropna()
    top10 = corr_with_y.abs().sort_values(ascending=False).head(10)
    if len(top10) > 0:
        plt.figure(figsize=(8,5))
        sns.barplot(x=top10.values, y=top10.index)
        plt.title("Top-10 | corr(feature, diagnosis_num)")
        plt.xlabel("|correlation|")
        plt.tight_layout(); plt.savefig("../results/plot3_bar_top10_corr.png", dpi=300); plt.show()
    else:
        print("[SKIP] Correlation bar: all correlations are NaN.")
else:
    print("[SKIP] Correlation bar: missing diagnosis_num or numeric features.")
```

```python
# Plot 4: Scatter (radius_mean vs texture_mean)
x_feat = pick_feature("radius_mean", ["perimeter_mean", "area_mean"]) 
y_feat = pick_feature("texture_mean", ["smoothness_mean", "compactness_mean"]) 
sub_scatter = df[[x_feat, y_feat, "diagnosis"]].dropna()
if len(sub_scatter) > 0:
    plt.figure(figsize=(7,5))
    sns.scatterplot(data=sub_scatter, x=x_feat, y=y_feat, hue="diagnosis", alpha=0.7)
    plt.title(f"{x_feat} vs {y_feat} (by diagnosis)")
    plt.tight_layout(); plt.savefig("../results/plot4_scatter_radius_vs_texture.png", dpi=300); plt.show()
else:
    print("[SKIP] Scatter: no rows available after dropping NaNs.")
```
